# RAG Corpus Analysis

Analysis of the 2,794 ArXiv abstracts in the knowledge base.

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from collections import Counter

# Load corpus
with open('../data/raw/arxiv_corpus.json', 'r') as f:
    papers = json.load(f)

print(f"Total papers: {len(papers)}")
```

## Category Distribution

```python
# Extract primary categories
categories = [p['categories'][0] if p.get('categories') else 'unknown' for p in papers]
cat_counts = Counter(categories)

# Top 15 categories
top_cats = cat_counts.most_common(15)
df_cats = pd.DataFrame(top_cats, columns=['Category', 'Count'])

plt.figure(figsize=(12, 6))
plt.barh(df_cats['Category'], df_cats['Count'])
plt.xlabel('Number of Papers')
plt.title('Top 15 ArXiv Categories in Corpus')
plt.tight_layout()
plt.show()

df_cats
```

## Publication Year Distribution

```python
# Extract years
years = [p['published'][:4] for p in papers if p.get('published')]
year_counts = Counter(years)

# Sort by year
df_years = pd.DataFrame(sorted(year_counts.items()), columns=['Year', 'Count'])

plt.figure(figsize=(12, 5))
plt.bar(df_years['Year'], df_years['Count'])
plt.xlabel('Year')
plt.ylabel('Number of Papers')
plt.title('Publication Year Distribution')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

df_years.tail(10)
```

## Abstract Length Statistics

```python
# Abstract lengths
lengths = [len(p['abstract']) for p in papers if p.get('abstract')]

print(f"Mean abstract length: {sum(lengths)/len(lengths):.0f} chars")
print(f"Median: {sorted(lengths)[len(lengths)//2]:.0f} chars")
print(f"Max: {max(lengths)} chars")
print(f"Min: {min(lengths)} chars")

plt.figure(figsize=(10, 5))
plt.hist(lengths, bins=50, edgecolor='black')
plt.xlabel('Abstract Length (characters)')
plt.ylabel('Frequency')
plt.title('Distribution of Abstract Lengths')
plt.axvline(sum(lengths)/len(lengths), color='red', linestyle='--', label='Mean')
plt.legend()
plt.tight_layout()
plt.show()
```
