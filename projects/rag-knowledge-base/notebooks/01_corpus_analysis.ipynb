# Corpus Analysis
Source: arXiv API (export.arxiv.org/api/query)

```python
import json
import pandas as pd
import numpy as np
from collections import Counter

with open('data/raw/arxiv_corpus.json') as f:
    corpus = json.load(f)
print(f'Loaded {len(corpus)} abstracts')
```

```python
cats = Counter(c for r in corpus for c in r['categories'])
pd.Series(cats).sort_values(ascending=False).head(10)
```

```python
import matplotlib.pyplot as plt
lengths = [len(r['summary'].split()) for r in corpus]
plt.figure(figsize=(10,4))
plt.hist(lengths, bins=50, edgecolor='black')
plt.title('Abstract Length Distribution')
plt.xlabel('Words')
plt.ylabel('Count')
plt.show()
```
