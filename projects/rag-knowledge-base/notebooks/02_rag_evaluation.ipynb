# RAG Evaluation

Evaluate retrieval quality and answer relevance.

```python
import sys
sys.path.insert(0, '..')

from src.rag_pipeline import RAGPipeline
import pandas as pd

# Initialize pipeline
pipeline = RAGPipeline()

# Test queries
test_queries = [
    "What is attention mechanism in transformers?",
    "How does few-shot learning work?",
    "What are diffusion models?",
    "Explain backpropagation in neural networks",
    "What is transfer learning?",
    "How do GANs generate images?",
    "What are the challenges in reinforcement learning?",
    "Explain self-supervised learning",
]

results = []
for q in test_queries:
    result = pipeline.query(q)
    results.append({
        'query': q,
        'answer_length': len(result.answer),
        'num_sources': len(result.sources),
        'retrieval_time_ms': result.retrieval_time_ms,
        'total_time_ms': result.total_time_ms,
        'strategy': result.strategy_used
    })

df = pd.DataFrame(results)
df
```

## Timing Analysis

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
x = range(len(df))
plt.bar(x, df['retrieval_time_ms'], label='Retrieval', alpha=0.7)
plt.bar(x, df['total_time_ms'] - df['retrieval_time_ms'], bottom=df['retrieval_time_ms'], label='Generation', alpha=0.7)
plt.xticks(x, [q[:30] + '...' for q in df['query']], rotation=45, ha='right')
plt.ylabel('Time (ms)')
plt.title('RAG Pipeline Timing Breakdown')
plt.legend()
plt.tight_layout()
plt.show()

print(f"Mean retrieval time: {df['retrieval_time_ms'].mean():.0f}ms")
print(f"Mean total time: {df['total_time_ms'].mean():.0f}ms")
```
