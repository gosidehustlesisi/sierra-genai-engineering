Live corpus from ArXiv, PubMed, and Wikipedia APIs.
**Sources:**
- ArXiv API: https://arxiv.org/help/api/index
- PubMed E-utilities: https://www.ncbi.nlm.nih.gov/books/NBK25500/
- Wikipedia REST API: https://www.mediawiki.org/wiki/API:Main_page

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_parquet('../data/processed/cleaned_documents.parquet')
print(f'Documents: {len(df)}')
print(df['category'].value_counts())
```

```python
# Text length distribution
df['text_len'] = df['text'].str.len()
df['clean_len'] = df['clean_text'].str.len()

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
sns.histplot(df['text_len'], bins=50, ax=axes[0])
axes[0].set_title('Raw Text Length')
sns.histplot(df['clean_len'], bins=50, ax=axes[1])
axes[1].set_title('Clean Text Length')
plt.tight_layout()
plt.show()
```

```python
# Category distribution by source
ct = pd.crosstab(df['source'], df['category'])
ct.plot(kind='bar', stacked=True, figsize=(10, 5))
plt.title('Documents by Source and Category')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
